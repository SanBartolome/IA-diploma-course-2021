## Realice aquí su gráfico animado
Se debe ver como en el documento TP4-indicaciones.pdf

```python
data = px.data.gapminder().query("continent == 'Europe' and pop > 2.e7")
unqiue_years = data["year"].unique()
unqiue_countries = data["country"].unique()

# Let's first separate the continent of a country
continent_per_country = data[["country", "continent"]].drop_duplicates().set_index("country") 

# Removing the continent
data = data[["year", "country", "pop"]].set_index(["year", "country"]) 
 
# Creating all possible "Year" "Country" combinations
data = data.reindex(pd.MultiIndex.from_product([unqiue_years, unqiue_countries]))  

# Cleanup
data = data.reset_index().rename(columns={"level_0": "year", "level_1": "country"})  

# Re-adding the continent
data = data.merge(continent_per_country, left_on="country", right_index=True)
```

```python
figure = px.bar(data, y = 'country', x = 'pop', text = 'pop', color = 'country', animation_frame="year", animation_group="country", orientation='h',\
            labels={"country":"","pop":"Population"}, range_x=[0,100000000])

#para colocar los números fuera de las barras
figure.update_traces(texttemplate='%{text:.2s}', textposition='outside')
figure.update_layout(xaxis_tickangle=-60, showlegend=False, \
                  title='Most populated countries in Europe by year',xaxis=dict(showticklabels=False))
figure
```
