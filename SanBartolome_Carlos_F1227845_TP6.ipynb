```python
from siphon.simplewebservice.ndbc import NDBC
import pandas as pd
import plotly.express as px

df = NDBC.realtime_observations('51002')
df.tail()
```

```python
series = df[['time','air_temperature','wind_speed','pressure']]
series.tail()
```

a) Revise la documentación (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.set_index.html) y modifique el dataframe *series* de manera que los índices sean igual a las fechas. 

```python
#código aquí
series.set_index('time', inplace=True)
series.tail()
```

b) Revise la documentación (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.rename_axis.html) y modifique el dataframe *series* de manera que los *headers* tengan la misma etiqueta *category*. 

```python
#código aquí
series.rename_axis('category', axis='columns', inplace=True)
series.tail()
```

c) En el tutorial de Timeseries se realizó el siguiente procedimiento:

```python
df = px.data.stocks(indexed=True)-1
fig = px.area(df, facet_col="company", facet_col_wrap=2)
fig.show()
```

Realice un procedimiento análogo para mostrar la variación de temperatura, rapidez del viento y presión en un mismo gráfico.

```python
#código aquí
fig = px.line(series, facet_col="category", facet_col_wrap=1)
fig.show()
```

d) Notará que, debido a la escala, no se puede visualizar bien la información. Revise la siguiente documentación: https://plotly.com/python/facet-plots/ y configure el gráfico para que los ejes sean independientes (revise el ejemplo de la sección *Facets With Independent Axes*).

```python
#código aquí
fig = px.line(series, facet_col="category", facet_col_wrap=1)
fig.update_yaxes(matches=None)
fig.show()
```

e) Agregue las siguientes líneas de código para cambiar los títulos de cada subplot y ocultar la leyenda (basado en https://stackoverflow.com/questions/65563922/how-to-change-subplot-title-after-creation-in-plotly):

```python
#código aquí (el mismo que utilizó en la parte "d")
fig = px.line(series, facet_col="category", facet_col_wrap=1)
fig.update_yaxes(matches=None)

fig.layout.annotations[2].update(text="Air temperature")
fig.layout.annotations[1].update(text="Wind speed")
fig.layout.annotations[0].update(text="Pressure")

fig.show()
```

f) Ahora deberá cambiar los nombres de los ejes. Para esto, revise la siguiente documentación: https://plotly.com/python/subplots/ (revise el ejemplo de la sección *Customizing Subplot Axes*)

```python
#código aquí
fig = px.line(series, facet_col="category", facet_col_wrap=1)
fig.update_yaxes(matches=None)

fig.layout.annotations[2].update(text="Air temperature")
fig.layout.annotations[1].update(text="Wind speed")
fig.layout.annotations[0].update(text="Pressure")

fig.update_yaxes(title_text="Temperature (°C)", row=3, col=1)
fig.update_yaxes(title_text="Speed (m/s)", row=2, col=1)
fig.update_yaxes(title_text="Pressure (Pa)", row=1, col=1)

fig.show()
```

g) Finalmente, oculte la leyenda, pues es redundante. Para averiguar cómo, googlee: "remove legend plotly"

```python
#código aquí
fig = px.line(series, facet_col="category", facet_col_wrap=1)
fig.update_yaxes(matches=None)

fig.layout.annotations[2].update(text="Air temperature")
fig.layout.annotations[1].update(text="Wind speed")
fig.layout.annotations[0].update(text="Pressure")

fig.update_yaxes(title_text="Temperature (°C)", row=3, col=1)
fig.update_yaxes(title_text="Speed (m/s)", row=2, col=1)
fig.update_yaxes(title_text="Pressure (Pa)", row=1, col=1)

fig.update_layout({"showlegend": False})

fig.show()
```
