```python
import pandas as pd
df = pd.read_csv('./BankChurners.csv')
df
```

```python
#Creating a Dataframe with the required columns from the base Dataframe
df1 = df[['Customer_Age', 'Gender']]
df1.head()
```

```python
#Assigning the population by age groups
bins = [25,35,40,45,50,55,60,75]
df1.loc[:,'Customer_Age'] = pd.cut(df1['Customer_Age'], bins=bins)
df1.head()
```

```python
#Creating the histogram
ax = pd.crosstab(df1['Customer_Age'],df1['Gender']).plot.barh(color={"F":"pink","M":"skyblue"})
```

```python
#Styling the histogram
ax.set_title("Male and female population by age group")
ax.set_xlabel("Population")
ax.set_ylabel("Age group")
ax.grid(color='gray', linestyle='-', linewidth=1, alpha=0.2)
ax.figure
```
